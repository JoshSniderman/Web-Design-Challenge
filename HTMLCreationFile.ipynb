```python
import pandas as pd
import numpy as np
```

```python
filepath = "Cities_Weather_Data.csv"
cities_df = pd.read_csv(filepath)
cities_df
```

```python
for row in cities_df.index:
    print('<tr>')
    print(f'''<th scope="row">{row}</th>''')
    for column in cities_df:
        print(f'''<td>{cities_df.loc[row, f"{column}"]}</td>''')
    print('</tr>')
```
